```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import time
```

```python
file = open(r"next_word_predictor.txt", "r", encoding = "utf8")
# Read the contents directly from the 'file' object
txt = file.read()

print(txt[:500])
```

```python
# Assign Tokens to every word in the text

tokenizer =  Tokenizer()
tokenizer.fit_on_texts([txt])
tokenizer.word_index
```

Word Index Length

```python
word_index_length = len(tokenizer.word_index)
print(f"The length of the word index is : {word_index_length}")
```

Generating Input Sequences

The input_sequences list contains sequences of tokens generated from the cleaned data.

Each sequence consists of a variable number of tokens, representing the context and target words for training a language model.

The printed output shows the first few input sequences as a demonstration.

```python
 #Splitting text into sentences on the basis of '\n' and tokenize input sequence
 for sentence in txt.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
  print(tokenized_sentence)
```

```python
# Convert Every sentence into input sequences
input_sequences = []
for sentence in txt.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1, len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

Finding Maximum Sequence Length

 calculates the maximum sequence length among all input sequences generated from the text data.

```python
max_length = max([len(x) for x in input_sequences])
max_length
```

Padding Input Sequences

pads the input sequences to ensure uniform length. It uses the pad_sequences function from TensorFlow Keras to pad the sequences with zeros (pre-padding) up to the maximum sequence length (max_len)

```python
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_length, padding = 'pre')
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
X
```

```python
y = padded_input_sequences[:,-1]
y
```

```python
X.shape
```

```python
y.shape
```

One-Hot Encoding Labels

one-hot encoding on the labels (y) to convert them into categorical format. It utilizes the to_categorical function from TensorFlow Keras to encode the labels into binary vectors

```python
y = to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
y.shape
```

```python
y
```

LSTM Language Model Architecture -




1.) Embedding Layer: This layer converts integer indices into dense vectors of fixed size. It takes input sequences of length 325 nd outputs dense vectors of size 100 for each word in the input sequence.

2.) LSTM Layer: This layer consists of 150 LSTM units. It takes the output of the embedding layer as input and processes the sequential data, capturing dependencies among words in the input sequence.

3.)Dense Layer: This layer is a fully connected (dense) layer with softmax activation function. It takes the output of the LSTM layer and predicts the probability distribution over all words in the vocabulary

```python
## Define the LSTM language model architecture
model = Sequential()
model.add(Embedding(4994,100,input_length=324))
model.add(LSTM(150))
model.add(Dense(4994,activation='softmax'))


## Compile the model
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.build(input_shape=(324, 100))
```

```python
 ##Summary of the Model
model.summary()
```

Training our model

We will train  defined LSTM language model (model) using the input features (X) and one-hot encoded labels (y) for a specified number of epochs (150) .During training, the model learns to predict the next word in a sequence based on the input context provided by the input sequences (X).

```python
# Train the LSTM language model
model.fit(X, y, epochs=150)
```

 Generating Text using Trained Language Model

```python
text = "Please let me know"

for i in range(25):
    # Tokenize the input text
    token_text = tokenizer.texts_to_sequences([text])[0]
    # Pad the tokenized text
    padded_token_text = pad_sequences([token_text], maxlen=max_length, padding='pre')
    # Predict the next word index
    pos = np.argmax(model.predict(padded_token_text))

    # Retrieve the word corresponding to the predicted index
    for word, index in tokenizer.word_index.items():
        if index == pos:
            # Append the predicted word to the input text
            text = text + " " + word
            print(text)
            # Simulate typing effect with a delay of 1 second
            time.sleep(1)
```
